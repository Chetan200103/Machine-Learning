```python
# Import libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sn

# Step 1: Data Preprocessing
# Clean column names
df.columns = df.columns.str.strip()

# Handle missing and incorrect values
df['EV_Sales_Quantity'] = pd.to_numeric(df['EV_Sales_Quantity'], errors='coerce').fillna(0)

# Select relevant features for clustering
df_filtered = df[['Year', 'EV_Sales_Quantity']].copy()

# Group by Year and Month for segmentation
grouped = df_filtered.groupby(['Year']).sum().reset_index()

# Step 2: Feature Scaling
scaler = StandardScaler()
scaled_features = scaler.fit_transform(grouped[['Year', 'EV_Sales_Quantity']])

# Step 3: Elbow Method to find optimal number of clusters
inertia = []
K_range = range(1, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_features)
    inertia.append(kmeans.inertia_)

# Plot elbow curve
plt.figure(figsize=(8, 5))
plt.plot(K_range, inertia, 'bo-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal K')
plt.grid(True)
plt.show()

# Step 4: Apply K-Means Clustering
optimal_k = 4  # You can choose this from the elbow plot
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
grouped['Cluster'] = kmeans.fit_predict(scaled_features)

# Step 5: Visualize the clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(data=grouped, y='EV_Sales_Quantity', hue='Cluster', palette='tab10')
plt.title('Clustered EV Sales by Month')
plt.xlabel('Month')
plt.ylabel('EV Sales Quantity')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()

# Step 6: Analyze each cluster
for cluster_id in range(optimal_k):
    cluster_data = grouped[grouped['Cluster'] == cluster_id]
    print(f"\nCluster {cluster_id} Statistics:")
    print(cluster_data.describe())
```
